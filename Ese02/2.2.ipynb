## <span style="color:green"> GRAFICO A) </span> Spazio discreto: reticolo cubico

Si dividono in $B=100$ blocchi da $100$, $10.000$ random walks (RW), ciascuno dei quali è una successione di 100 passi, con direzione e verso casuali, in un reticolo cubico 3D. Si è scelto di far cominciare ogni RW dall'origine, con passo $a=\pm 1$, orientato lungo uno degli assi x, y o z.

Per ogni blocco si è calcolata la media della distanza dall'origine a passo $N$ fissato: $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$.

Si riportano nel grafico seguente i valori medi totali, che considerano tutti i blocchi, con le loro deviazioni standard della media.  

Infine si confronta la curva con la funzione del tipo $f(N)=k\sqrt{N}$, arrivando alla conclusione che il fit migliore si ottiene per $k=1$.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Definisci la funzione di fit
def fit_function(x, k):
    return k * np.sqrt(x)

# Carica i dati dal file
data = np.loadtxt('2.RW_Discreto.out')
x = np.arange(1, 101)

# Esegui il fit dei dati
params, covariance = curve_fit(fit_function, x, data[:, 0])

# Estrai il parametro k dalla lista dei parametri
k_best_fit = params[0]

# Calcola le y fitte utilizzando il parametro k migliore
y_fit = fit_function(x, k_best_fit)

# Plotta i dati e la curva di fit
plt.errorbar(x, data[:, 0], yerr=data[:, 1], color='green', label='Dati della simulazione')
plt.plot(x, y_fit, 'r--', label=f'Fit: k = {k_best_fit:.2f}')
plt.xlabel("Numero di step dall'origine")
plt.ylabel(r'Distanza media: $\frac{1}{B}\sum_{i}^B\sqrt{\langle|r_N|^2\rangle_i}$')
plt.grid(True)
plt.title('A) RW reticolo cubico discreto')
plt.legend()
plt.show()

# Stampa il parametro k migliore
print(f'Il parametro k migliore è: {k_best_fit:.2f}')
```

```python
plt.errorbar(x, data[:,0] - np.sqrt(x), yerr=data[:,1], color='green', fmt='o', markersize=5)
plt.xlabel("Numero di step dall'origine")
plt.ylabel(r'$\sqrt{ \langle|r_N|^2 \rangle }-\sqrt{N}$')
# Disegna una retta orizzontale sul valore atteso
plt.axhline(y=0, color='r', linestyle='--', label="Valore atteso")
plt.grid(True)
plt.title('A) Discrepanze rispetto al fit Vs incertezze statistiche (Dev. St. Media): comportamento diffusivo')
plt.legend()
plt.show()
```
