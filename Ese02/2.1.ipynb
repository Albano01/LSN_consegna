## <span style="color:blue"> GRAFICO A) </span> Integrale con densità di probabilità uniforme

Si vuole risolvere l'integrale 1D $$I =\int_0^1\frac{\pi}{2}\cos(\pi x/2) dx = 1 \hspace{1cm}\hbox{con}\hspace{1cm} f(x) = \frac{\pi}{2}\cos(\pi x/2)$$
con Metodo Montecarlo, dapprima estraendo $x$ da una distribuzione uniforme $\in [0,1]$.
Nel grafico vengono riportati, con il metodo della somma a blocchi, i risultati per $M=10.000$ estrazioni di $x$ per blocco, in $N=100$ blocchi.

```python
import matplotlib.pyplot as plt
import numpy as np

# Vettore di numeri ordinati da 1 a 100 (infatti il primo blocco contiene già 10000 numeri, non ha senso partire da 0!)
x = np.arange(1, 101) 

# Carica i dati dal file
data = np.loadtxt('1.I_uniforme.out')

x*=10000  # Tiri per blocco

# Disegna una retta orizzontale sul valore atteso
plt.axhline(y=0, color='r', linestyle='--', label="Valore atteso")

plt.errorbar(x,data[:,0]-1,yerr=data[:,1], color='blue', label="Dev St della media")
plt.xlabel('Numero di tiri')
plt.ylabel('Integrale -1')
plt.grid(True)

plt.title('A) Integrale con densità di probabilità uniforme')

plt.legend()
plt.show()
```

## <span style="color:green"> GRAFICO B) </span> Integrale con densità di probabilità simile all'integranda

Scelta come densità di probabilità $$p(x)=\frac{3}{2}(1-x^2)$$funzione positiva e normalizzata nell'intervallo $[0,1]$ che somiglia all'integranda $f(x)$, la sua comulativa è $P(x)=\frac{3}{2}(x-\frac{x^3}{3})$. L'inversa di nostro interesse risulta quindi essere: $$x=-2\cdot sin\bigg(\frac{arcsin(-y)}{3}\bigg)$$ tale che con y estratto uniformemente in $[0,1]$ essa sia proprio la generatrice di una distribuzione che va come $p(x)$.

Si riportano nel grafico i risultati ottenuti con le stesse modalità (somma a blocchi) del caso precedente.

```python
# Vettore di numeri ordinati da 1 a 100 (infatti il primo blocco contiene già 10000 numeri, non ha senso partire da 0!)
x = np.arange(1, 101) 

# Carica i dati dal file
data2 = np.loadtxt('1.I_NONuniforme.out')

x*=10000 # Tiri per blocco

# Disegna una retta orizzontale sul valore atteso
plt.axhline(y=0, color='r', linestyle='--', label="Valore atteso")

plt.errorbar(x,data2[:,0]-1,yerr=data2[:,1], color='green', label='Dev St della media')
plt.xlabel('Numero di tiri')
plt.ylabel('Integrale -1')
plt.grid(True)

plt.title(r'B) Integrale con densità di probabilità $p(x) = \frac{3}{2}(1-x^2)$')

plt.legend()
plt.show()
```

## <span style="color:orange"> GRAFICO C) </span> Confronto tra i due grafici precedenti
Per semplicità di visualizzazione si pongono a confronto i grafici precedenti: si noti come, se si sceglie una densità di probabilità più simile all'integranda, l'incertezza statistica si riduce notevolmente (in tal caso di un fattore 10).

```python
# Vettore di numeri ordinati da 1 a 100 (infatti il primo blocco contiene già 10000 numeri, non ha senso partire da 0!)
x = np.arange(1, 101) 

x*=10000 # Tiri per blocco


# Carica i dati dal file
data = np.loadtxt('1.I_NONuniforme.out')

plt.errorbar(x,data[:,0]-1,yerr=data[:,1], color='green', label='Densità non uniforme')

# Carica i dati dal file
data1 = np.loadtxt('1.I_uniforme.out')

plt.errorbar(x,data1[:,0]-1,yerr=data1[:,1], color='blue', label='Densità uniforme')

plt.xlabel('Numero di tiri')
plt.ylabel('Integrale -1')
plt.grid(True)

plt.title('C) Confronto')

plt.legend()
plt.show()
```
