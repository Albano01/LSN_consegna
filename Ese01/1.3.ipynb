## <span style="color:blue"> GRAFICO A) </span> Stima di pigreco $\pi$
Sia dato un tavolo con linee parallele equidistanziate a distanza $d$. Qual è la probabilità che lanciando casualmente un ago di lunghezza $L<d$ questo intersechi una delle linee?
$$P = \frac{2L}{\pi d} \Rightarrow \pi = \frac{2L}{Pd} = \lim_{N_{tiri} \to \infty}\frac{2L\cdot N_{tiri}}{d\cdot N_{intersezioni}}$$
Usando il metodo della media della somma progressiva di blocchi, si sono presi $100$ blocchi, ognuno con $100$ stime di $\pi$, date ciascuna da $N_{tiri}=10.000$ lanci casuali dell'ago. L'incertezza statistica è resa dalla deviazione standard della media.

Nota: si è preso $d=4$ ed $L=1$ così che $d>L$, ma NON $d\gg L$, che avrebbe causato $P\ll 1$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Carica i dati dal file
data = np.loadtxt('3.Pigreco.out')

x = np.arange(100)

plt.errorbar(x+1,data[:,0],yerr=data[:,1], color='blue')

# Aggiungi la riga tratteggiata orizzontale al valore di pi
plt.axhline(np.pi, color='red', linestyle='--', label='Valore atteso ($\pi$)')

plt.xlabel('Numero blocchi sommati')
plt.ylabel('Pigreco')
plt.grid(True)
plt.title('A) Esperimento di Buffon: stima di pigreco')
plt.legend()
plt.show()
```
